### Dados mortes e pacientes recuperados:

Vou deixar os dados para fins didáticos, porém o DF de recuperados está praticamente todo com valor zero.

```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv", sep=',')
```

```python
# metodo display do notebook, para mostrar dois dataframes seguidos:

display(deaths.sample(5))
```

```python
def latest_by_country(data):
  return data.groupby('Country/Region').sum().iloc[:, -1]
  # iloc para pegar apenas a ultima coluna, referente ao dia mais atual
```

```python
informations = [
                latest_by_country(confirmed),
                latest_by_country(deaths),
]
```

```python
display(latest_by_country(confirmed))
display(latest_by_country(deaths))
```

```python
combined = pd.concat(informations, axis=1)
```

```python
combined.columns = ['confirmed', 'deaths']
```

```python
combined
```

### Taxa de letalidade:

A taxa de letalidade é expressa em %.

O número de casos letais dentro dos casos que foram letais e não letais.

Por exemplo, imagine uma doença que em janeiro/22 afetou 100 pessoas e dessas, 5 morreram. A letalidade foi de 5% se a doença deixou de existir pela historia da humanidade.

Com o passar do tempo, a doença e os tratamentos evoluem, e a taxa de letalidade muda.

```python
sum_up = combined.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed']
print(letality_rate_1.round(4) * 100)
```

Nesta primeira estimativa da taxa de letalidade, dividimos as mortes pelos casos confirmados.

1.13% das pessoas dos casos confirmados do conjunto de dados que estamos utilizando, morreram.

### Taxa de letalidade por pais, e a dificuldade de resumir informações

```python
letality_rate_1 = combined['deaths'] / combined['confirmed'] * 100
combined['letality_rate_1'] = letality_rate_1
combined.sample(10)
```

20 paises com a maior taxa de letalidade:

```python
combined.sort_values('letality_rate_1', ascending=False).head(20)  #.query('confirmed > 1000').head(10)
```

# Importando dataset UN world population:

These CSV files are encoded in UTF-8 and all of them have the following columns:

- SortOrder (numeric): record counter
- LocID (numeric): numeric code for the location; for countries and areas, it follows the ISO 3166-1 numeric standard
- Notes (string): symbol linked to location notes file (available for download below)
- ISO3_code (string): ISO 3166-1 alpha-3 location codes
- ISO2_code (string): ISO 3166-1 alpha-2 location codes
- SDMX_code (string): SDMX Global Registry, cross domain code list for geographical areas (version 2.0) represents a combination of reference area codes in M49 and ISO-3166 classification for international data exchange and interoperability https://sdmx.org/?page_id=3215 and https://registry.sdmx.org/ws/public/sdmxapi/rest/codelist/SDMX/CL_AREA/2.0
- LocTypeID (numeric): code for location type
- LocTypeName (string): type of location
- ParentID (numeric): numeric code of the parent location
- Location (string): name of the region, subregion, country or area
- VarID (numeric): numeric code for the scenario
- Variant (string): projection scenario name (Medium is the most used); for more information see Definition of Projection Scenarios
- Time (numeric): year the data refers to
- MidPeriod (numeric): numeric value identifying the mid period of the data, with the decimal representing the month (e.g. 1950.5 for July of 1950)

```python
un_population = pd.read_csv("https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/CSV_FILES/WPP2022_TotalPopulationBySex.zip")
```

```python
un_population.head()
```

```python
# pegando registros apenas do Brasil
un_population.query('Location == "Brazil"')
```

```python
# multiplicando populações por 1000 para ter valores das populações reais
un_population[['PopMale', 'PopFemale', 'PopTotal']] = un_population[['PopMale', 'PopFemale', 'PopTotal']] * 1000
```

```python
# Verificando registros do Brasil com população atualizada
un_population.query('Location == "Brazil"')
```

```python
# Investigando o pq que tem tantas linhas com Brasil. Descobri que é a coluna 'Variant':
un_population.query('Time==2022')[['LocID', 'Variant', 'Time', 'MidPeriod', 'Location', 'PopTotal']].set_index('Location').loc['Brazil']
```

Variant (string): projection scenario name (Medium is the most used); for more information see Definition of Projection Scenarios

```python
# Utilizando 'medium' conforme indicado acima e colocando em uma variavel
un_population_total = un_population.query('Time==2022 and Variant=="Medium"')[['Location', 'PopTotal']].set_index('Location')
```

```python
un_population_total.sample(10)
```

```python
# Juntando dataframes
combined.join(un_population_total)
```

```python
# filtrando por casos confirmados para ver se o join deu certo (pelo visto nao, como é o caso do US)
# os nomes dos paises certamente nao estão batendo
combined.join(un_population_total).sort_values('confirmed', ascending=False)
```

```python
un_population_total.index
```

```python
# query com situação especifica 'contains' usando engine/parser do python
un_population.query('Location.str.contains("United States of America")', engine='python')
```

```python
# setando Time==2022 and Variant==Medium
# usar esta célula para localizar paises onde 'location' não bate

def locate(population, country):
  return population.query(f'Location.str.contains("{country}") and Time==2022 and Variant=="Medium"', engine='python')

locate(un_population, 'Viet')
```

```python
# função para renomear USA para US utilizando função .map()

def rename_location(location):
  if location == 'United States of America':
    return 'US'
  if location == 'Republic of Korea':
    return 'Korea, South'
  if location == 'Russian Federation':
    return 'Russia'
  if location == 'Türkiye':
    return 'Turkey'
  if location == 'Iran (Islamic Republic of)':
    return 'Iran'
  if location == 'China, Taiwan Province of China':
    return 'Taiwan*'
  if location == 'Viet Nam':
    return 'Vietnam'
  return location

un_population_total['Location_hopkins'] = un_population_total.index.map(rename_location)
```

```python
un_population_total.query('Location=="United States of America"')
```

```python
un_population_total = un_population_total.set_index('Location_hopkins')
```

```python
un_population_total.head()
```

```python
un_population_total.query('Location_hopkins=="US"')
```

```python
# Países com maior número de casos confirmados
combined_expanded = combined.join(un_population_total).sort_values('confirmed', ascending=False)
combined_expanded.head(30)
```

## Taxa de incidencia e de mortalidade

```python
combined_expanded['incidence_ratio'] = combined_expanded['confirmed'] / combined_expanded['PopTotal'] * 100000
combined_expanded['mortality_rate'] = combined_expanded['deaths'] / combined_expanded['PopTotal'] * 100000
combined_expanded.head(30)
```

Incidence/mortality ratio: Número de pessoas infectadas/mortas por 100mil habitantes.

```python
# Paises com mortalidade mais alta
combined_expanded.sort_values('mortality_rate', ascending=False).head(30)
```

```python
# Paises com incidencia mais alta
combined_expanded.sort_values('incidence_ratio', ascending=False).head(30)
```

```python
# Fazendo novo dataframe com dados somados worldwide
# .T transposição
grouped_per_day = pd.DataFrame([confirmed.sum()[3:], deaths.sum()[3:]],
                               index=['confirmed', 'deaths']).T
grouped_per_day.head()
```

```python
# letalidade worldwide
grouped_per_day['letality_rate'] = (grouped_per_day['deaths'] / grouped_per_day['confirmed']) * 100
grouped_per_day.head()
```

```python
plt.figure(figsize=(20, 10))

ax = grouped_per_day['letality_rate'].plot()

# plt.legend(['Mortes por casos confirmados', 'Mortes / (Mortes + Recuperados'])  # nao temos recuperados mais
plt.title('Mortes por casos confirmados no mundo todo', fontsize=25)
ax.set_xlabel('Fonte: dados Johns Hopkins em https://github.com/CSSEGISandData/COVID-19', fontsize=20)
ax.set_ylabel('% letalidade', fontsize=20)
plt.grid()
```

```python
sns.boxplot(tmdb.vote_average, width=0.3)
```

Essas são maneiras diferentes de visualizarmos as informações do nosso conjunto, cada uma com suas propriedades. Por exemplo, no histograma, apesar de conseguirmos identificar as frequências com mais exatidão, é impossível visualizar a mediana ou os quartis, que são algumas das principais informações no boxplot. Dessa forma, a utilização de cada visualização vai depender dos nossos objetivos.

# **Analisando também o movielens**

Nós analisamos visualmente a média dos filmes do TMDB 5000, e percebemos que seu comportamento é parecido com uma distribuição normal, com exceção do lado esquerdo do gráfico.

Agora, vamos comparar esse conjunto com os dados do MovieLens, procurando saber se esse comportamento se repete. Primeiramente, carregaremos o arquivo ratings.csv e o importaremos, atribuindo essa leitura a uma variável notas. Feito isso, exibiremos os 5 primeiros registros desse dataset:

```python
notas = pd.read_csv('ratings.csv')
notas.head()
```

Temos aqui um formato diferente, com a identificação do usuário (userId), a identificação do filme (movieId) e a nota que foi dada para cada filme (rating).

Ou seja, não temos as médias de cada filme, então teremos que calculá-la. Para isso, agruparemos as notas por filme (notas.groupby("movieId")) e tiraremos a média apenas do campo rating (o único em que essa medida faz sentido), atribuindo o retorno a uma variável nota_media_por_filme. Então, exibiremos os 5 primeiros registros:

```python
nota_media_por_filme = notas.groupby('movieId').mean()['rating']
nota_media_por_filme.head()
```

Em seguida, geraremos a primeira visualização dessas médias. Para isso, repetiremos o processo que fizemos para as médias do TMDB 5000, acrescentando .values à variável nota_media_por_filme de modo a pegar somente os valores, e não os ids de cada filme:

```python
ax = sns.distplot(nota_media_por_filme.values)

ax.figure.set_size_inches(12, 6)
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no MovieLens')
```

Dessa vez, temos um gráfico com notas de 0 a 5. Novamente, ainda que não tenhamos filmes com média 0, temos alguns cuja média é 5. Portanto, é de se esperar que existam alguns filmes com poucos votos.

Com notas.groupby("movieId").count(), contaremos quantos votos cada um dos filmes possui nesse conjunto, atribuindo o resultado a uma variável quantidade_de_votos_por_filme. Com ela, faremos uma query() que selecionará apenas os filmes com 10 ou mais votos.

```python
quantidade_de_votos_por_filme = notas.groupby('movieId').count()
quantidade_de_votos_por_filme.query('rating >= 10')
```

Quero, então, saber quais são os filmes que compõem esse conjunto. Para isso, utilizaremos o index (que se refere a movieId), já que esse campo é um índice e não uma coluna determinada. Criaremos, com essa query(), uma variável filmes_com_pelo_menos_10_votos.

```python
quantidade_de_votos_por_filme = notas.groupby('movieId').count()
filmes_com_pelo_menos_10_votos = quantidade_de_votos_por_filme.query('rating >= 10').index
filmes_com_pelo_menos_10_votos
```

O formato devolvido por essa função é Int64Index. Podemos extrair somente os valores, criando um array, com filmes_com_pelo_menos_10_votos.values.

A variável nota_media_por_filme, que criamos anteriormente, é uma série que contém as médias e os ids dos filmes no conjunto. Agora, queremos somente os valores cujo índice está contido em filmes_com_pelo_menos_10_votos.

Para isso, usaremos o indexador .loc[]. Como parâmetro, ele pode receber um índice ou até mesmo um array, que é exatamente o que temos em filmes_com_pelo_menos_10_votos! Atribuiremos a series retornada pelo .loc[] a uma nova variável, de nome bastante extenso mas igualmente explicativo:



```python
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos.values]
nota_media_dos_filmes_com_pelo_menos_10_votos.head()
```

Agora, podemos gerar novamente um histograma, agora com dados mais significantes:

```python
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos)

ax.figure.set_size_inches(12, 6)
ax.set(xlabel='Nota Média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no Movielens')
```

Se traçarmos uma "mediana imaginária" por volta do 3,5, teremos uma curva um pouco mais abaulada à esquerda, e um pouco mais inclinada à direita - da mesma forma que no histograma do TMBD 5000. Claro, podem existir outros fatores que tornam a distribuição daquele conjunto mais "espalhada".

Vamos gerar também o boxplot dos dados do MovieLens:

```python
ax = sns.boxplot(x=nota_media_dos_filmes_com_pelo_menos_10_votos.values, width=0.3)

ax.figure.set_size_inches(12, 6)
ax.set(xlabel='Nota Média do filme')
ax.set_title('Distribuição de nota méédia dos filmes no Movielens')
```

Repare que essa nova visualização também se assemelha àquela do TMDB, com os quartis à esquerda (25% e 50%) mais densos que os da direita.

Uma análise "visual" dessas distribuições estão nos indicando que as pessoas se comportam de maneira similar. A seguir, continuaremos explorando esses dados.

# **video_03**

Nós fomos capazes de analisar, de forma visual, as distribuições das médias de notas atribuídas a filmes em dois sites diferentes. Também aprendemos a extrair algumas informações do boxplot, como a mediana e o comportamento das médias - por exemplo, 50% das notas são inferiores a 3,5, e 50% são superiores.

Porém, é difícil afirmarmos algo sobre um filme específico nesse conjunto. Por exemplo, supondo que o filme Toy Story tenha uma nota 3,9... o que isso quer dizer sobre esse filme? É um bom filme? Parece que sim... mas quão bom?

E se o determinado filme tiver uma média 3, ele é bom ou ruim? Precisamos entender, agora, como uma nota específica se encaixa dentro dessa distribuição.

Uma maneira de fazermos isso é acumulando a quantidade de filmes com uma determinada média conforme a gradação vai aumentando no histograma. Assim, se tivermos 5 filmes com notas entre 2 e 2.1, por exemplo, e depois 10 filmes com 2.1 e 2.2, teríamos 15 filmes nessa última gradação - e assim sucessivamente até o total de filmes.

Na documentação do distplot() do Seaborn, não existe uma maneira direta de indicarmos essa gradação cumulativa. Porém, o distplot() chama duas funções do matplotlib (hist() e kde()), e pode receber argumentos extras por meio de hist_kws e kde_kws. Dito isso, a função hist() do matploblib tem uma opção cumulative!

Portanto, na função displot(), passaremos tanto hist_kws quanto kde_kws recebendo {'cumulative':True}:

```python
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos,
                  hist_kws={'cumulative':True},
                  kde_kws= {'cumulative':True})

ax.figure.set_size_inches(12, 6)
ax.set(xlabel='Nota média', ylabel='Proporção acumulada de filmes')
ax.set_title('Média de votos em filmes no Movielens')
```

No eixo y, ao invés da densidade, temos agora a proporção acumulada dos filmes, que nos permite estipular as porcentagens nesse conjunto.

Por exemplo, podemos perceber que aproximadamente 50% dos filmes tem uma nota média menor que 3,5. Se o filme Toy Story tem uma média 3,9, ele está acima de 80% dos filmes do conjunto. Já um filme com nota 2,5 estaria abaixo de 90% dos filmes!

Vamos repetir esse processo para os dados do TMDB 5000:

```python
ax = sns.distplot(tmdb_com_mais_de_10_votos.vote_average, 
                  hist_kws = {'cumulative':True}, 
                  kde_kws = {'cumulative':True})

ax.figure.set_size_inches(12, 6)
ax.set(xlabel='Nota média', ylabel='Proporção acumulada de filmes')
ax.set_title('Média de votos em filmes no TMDB 5000')
```

Com isso, conseguimos as distribuições cumulativas das notas médias dos nossos filmes. Interessante, não?

Continuando com as operações que estamos fazendo com o nosso conjunto de dados, vamos analisar a média das médias (isso mesmo!) dos filmes do MovieLens:

```python
nota_media_dos_filmes_com_pelo_menos_10_votos.mean()
```

Sabendo do fato de que a nota mínima no MovieLens é 0,5 e a máxima é 5, poderíamos esperar que a média das médias fosse 2.75. Portanto, em geral, parece que os usuários atribuem notas maiores que 2.75 aos filmes.

Mas vamos pensar em alguns outros cenários. Imagine que, por exemplo, a família do nosso instrutor Guilherme é composta por três pessoas. Certo dia, logo após assistirem um filme muito ruim, todos eles dão nota 1 a ele, o que resulta em uma média 1 - muito diferente da média que estipulamos para o conjunto, não é?

É importante notarmos que cada um dos registros dessa amostra não é independente dos outros, afinal, todas essas pessoas fazem parte da mesma família, assistiram o filme juntas e deram a nota imediatamente após assistirem o filme. Portanto, esses dados são muito tendenciosos.

Portanto, temos que tomar cuidado não só com independência da amostra e outras características dos nossos dados, mas também com o tamanho da amostra.

Vamos começar exibindo a média dos 5 primeiros registros:

```python
nota_media_dos_filmes_com_pelo_menos_10_votos[0:5].mean()
```

Em seguida, mostraremos o tamanho do nosso conjunto com len():

```python
len(nota_media_dos_filmes_com_pelo_menos_10_votos)
```

Podemos, então, criar um for que abranja todo o nosso conjunto:

```python
# for i in range(1, len(nota_media_dos_filmes_com_pelo_menos_10_votos)):
```

Depois, calcularemos as médias resultantes da entrada de cada registro no conjunto - ou seja, a média do primeiro elemento, dos dois primeiros elementos, dos três primeiros elementos, e assim por diante.

```python
# for i in range(1, len(nota_media_dos_filmes_com_pelo_menos_10_votos)):
#     nota_media_dos_filmes_com_pelo_menos_10_votos[0:i].mean()
```

Criaremos uma variável medias que receberá como argumento uma lista (list()). Com medias.append(), passaremos o retorno da nossa iteração para essa variável, e a exibiremos na tela:

```python
medias = list()
for i in range(1, len(nota_media_dos_filmes_com_pelo_menos_10_votos)):
    medias.append(nota_media_dos_filmes_com_pelo_menos_10_votos[0:i].mean())
medias
```

Vamos plotar essas médias em um gráfico para que possamos analisar o seu comportamento. Para isso, importaremos o matlplotlib e passaremos medias para a função plt.plot():

```python
import matplotlib.pyplot as plt

medias = list()
for i in range(1, len(nota_media_dos_filmes_com_pelo_menos_10_votos)):
    medias.append(nota_media_dos_filmes_com_pelo_menos_10_votos[0:i].mean())
plt.plot(medias)
```

Perceba que, no começo, a média tem quedas e elevações constantes. Porém, em determinado momento do conjunto (por volta dos 1000 registros), a distribuição parece se estabilizar, mantendo-se quase constante.

Isso acontece pois, quando temos poucos elementos, cada novo registro no conjunto faz uma diferença bastante significante. Porém, conforme a amostra vai se tornando mais populada, os novos valores passam a interferir menos no resultado final.

Isso significa que, se tivéssemos uma amostra muito pequena, ou mal escolhida, poderíamos ter chegado a resultados diferentes e que levassem a outras interpretações a respeito desses dados. Com 580 filmes, por exemplo, temos uma média de 3.52, um valor mais alto que a média final de 3.43.

Repare, também, que esses filmes estão em uma ordem específica, fornecida pelo próprio dataset, e não sabemos se essa ordem tem alguma característica oculta que possa estar afetando essa distribuição. Vamos testar isso aleatorizando a ordem do nosso conjunto. Para isso, importaremos o numpy (import numpy as np) e passaremos uma seed aleatória ("semente", em inglês) que nos permita fixar uma distribuição (de modo que o processo seja reproduzível).

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(75243)
temp = nota_media_dos_filmes_com_pelo_menos_10_votos.sample(frac=1)  
# sample para aleatoriezar e testar se a ordem do dataset interfere nos resultados

medias = list()
for i in range(1, len(temp)):
    medias.append(temp[0:i].mean())
plt.plot(medias)
```

Dessa vez, nosso gráfico começa com valores altos, diminui bastante e então progride de maneira mais estável, tornando-se muito menos sensível à entrada de novas médias.

Esses testes visam demonstrar que não podemos ter certeza de que a média de um conjunto de dados representa adequadamente o mundo real. Antes de finalizarmos o processo, vamos refatorar esse código.

Entretanto, no início do gráfico, quando temos poucas amostras no conjunto, essa média varia substancialmente. Vamos analisar como isso influiencia nos nossos testes, calculando o Teste Z para cada um dos valores da amostra:

```python
# medias = [calcula_teste(i) for i in range(1, len(temp))]
```

Criaremos, então, uma função calcula_teste(i). Nela, colocaremos o código referente à media do trecho analisado (temp[0:i].mean()). Além disso, aplicaremos o ztest() para essas notas, estipulando como referência o valor 3.4320503405352603. Essa função nos dará dois retornos: stat e p.

```python
np.random.seed(75241)
temp = notas1.sample(frac=1)['rating']

def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i], value = 3.4320503405352603) 
    return (media, p)

medias = [calcula_teste(i) for i in range(2, len(temp))]  # a partir do 2 elemento pois nao da pra calcular p value só com 1 amostra
medias
```

Essa função nos devolverá diversas duplas contendo a média e o p value (exceto a primeira, já que não é possível calcular o p value com apenas um valor).

Consertaremos esse pequeno problema alterando a range() da função para que ela se inicie a partir do segundo elemento. Além disso, em return, adicionaremos i, o que nos permitirá acompanhar o número de amostras que gera cada média:

```python
def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i], value = 3.4320503405352603) 
    return (i, media, p)

medias = [calcula_teste(i) for i in range(2, len(temp))]
medias
```

Portanto, em nossas análises estatísticas, devemos prestar atenção não só ao p value, como também ao intervalo de confiança e a como esses valores se aplicam à situação com a qual estamos trabalhando.

Como é que podemos garantir que os dados das nossas notas se comportam como uma distribuição normal? Existe testes estatísticos criados exatamente para isso! Existem certas vantagens e características que devem ser consideradas na escolha de um desses testes. No nosso caso, utilizaremos o normaltest() do módulo scipy.stats.

Importaremos o normaltest(), que pode receber os dados (notas1.rating) e outras características opcionais. Ele nos devolve a estatística (stats) e o p value, que no caso é o valor que estamos interessados.

```python
from scipy.stats import normaltest
```

```python
stat, p = normaltest(notas1.rating)
p
```

Teremos um p value de 0.00011053430732728716 - ou seja, menor do que 0.05. Como essa função testa a hipótese nula de que o conjunto de dados venha de uma distribuição normal, um resultado menor que 0.05 indica que não estamos trabalhando com uma distribuição normal.

Isso também significa que não poderíamos aplicar o Teste T ou o Teste Z, mas sim testes que funcionam com outros tipos de distribuição. Nos cursos de estatística da Alura, abordamos testes que funcionam nessas situações: os testes não paramétricos.

```python
_, p = normaltest(notas1.rating)  # o underline indica que o valor a em (a, b) que voltam no retorno da função pode ser descartado
p
```

Para compararmos os conjuntos que separamos anteriormente (por exemplo, notas1 e notas593), utilizaremos a função ranksums(), que também é encontrada no módulo scipy.stats. Essa função consiste no teste Wilcoxon, que faz uma análise baseada em ranqueamento.

Depois de importamos essa função, vamos chamá-la passando como parâmetros os conjuntos notas1.rating e notas593.rating. Armazenaremos o p value retornado em uma variável p, que imprimiremos na tela:

```python
from scipy.stats import ranksums
```

```python
_, p = ranksums(notas1.rating, notas593.rating)
p
```

Nesse caso, o p value é significativo, descartando a hipótese nula. No ranksums(), a hipótese nula é de que as duas amostras vieram da mesma distribuição. Já a hipótese alternativa, que admitiremos nesse caso (já que o p value foi bem pequeno), é a de que os valores de uma amostra têm uma tendência a serem maiores que os da outra amostra.

Repare que, quando trabalhamos com Teste T e Teste Z estávamos nos referindo às médias - o que fazia sentido em uma distribuição normal. Porém, em uma distribuição não paramétrica, não usamos as médias, mas sim amostras aleatórias de cada conjunto testado.

Ou seja, se pegarmos notas aleatórias para Toy Story e Silence of the Lambs, as notas deste último, em geral, são maiores. Análises como essa, junto com as médias, os boxplots, os histogramas e outras informações que aprendemos a testar anteriormente, nos permitem concluir quando um filme é melhor que outro.

## Construindo a árvore de decisão

Então, como ele funciona? Ao invés de eu ter o meu conjunto inteiro de dado, aquelas quase 285 mil transações - e desse conjunto eu pegar 10% para teste e 90% para treino - eu vou fazer diferente.

[01:38] Imagine só se temos esse conjunto inteiro e agora eu quero pegar 10% desses dados para o teste. Mas desse conjunto inteiro, só 400 e poucas transações são aquelas transações com fraude, então tenho uma chance muito alta de selecionar esse meu conjunto de teste e não pegar nenhuma transação que seja fraude.

[02:01] Então não vou conseguir validar o meu modelo direito, porque o modelo tem que ter os dois tipos de transações - tanto no conjunto de treino, como um conjunto de teste. Então com o StratifiedShuffleSplit, ao invés de fazermos essa divisão do conjunto inteiro, ele faz diferente.

[02:18] Vamos imaginar que de um lado eu tenho as minhas transações com fraude e do outro eu tenho as minhas transações sem fraude. Então ele vai começar nas minhas transações com fraude, eu vou pegar 10% delas para o teste e 90% delas para o treino.

[02:34] Vou para as minhas transações sem fraude, pego 10% delas para o teste e 90% delas para o treino. Entendeu como funciona? Assim eu vou garantir que o meu conjunto vai ter 10% das transações de cada uma das classes.

```python
from sklearn.model_selection import StratifiedShuffleSplit

def executar_validador(X, y):
  validador = StratifiedShuffleSplit(n_splits=1, 
                                    test_size=0.1, 
                                    random_state=0)

  for treino_id, teste_id in validador.split(X, y):
    X_train, X_test = X[treino_id], X[teste_id]
    y_train, y_test = y[treino_id], y[teste_id]
  
  return X_train, X_test, y_train, y_test
```

```python
# %% time mostra o tempo que levou pra executar esta célula
%%time

from sklearn import tree

def executar_classificador(classificador, X_train, X_test, y_train):
  arvore = classificador.fit(X_train, y_train)
  y_pred = arvore.predict(X_test)

  return y_pred
```

```python
# execução do validador

X = dados.drop('Class', axis=1).values
y = dados['Class'].values

X_train, X_test, y_train, y_test = executar_validador(X, y)
```

```python
#execução do classificador DecisionTreeClassifier
classificador_arvore_decisao = tree.DecisionTreeClassifier()
y_pred_arvore_decisao = executar_classificador(classificador_arvore_decisao,
                                               X_train,
                                               X_test, 
                                               y_train)
```

## Medindo a acurácia, precisao, recall e matriz de confusão da árvore

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix

def validar_arvore(y_test, y_pred):
  print(f'Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%')
  print(f'Precision: {precision_score(y_test, y_pred) * 100:.2f}%')
  print(f'Recall: {recall_score(y_test, y_pred) * 100:.2f}%')
  print()
  print('Matriz de confusão')
  print(confusion_matrix(y_test, y_pred))
```

```python
validar_arvore(y_test, y_pred_arvore_decisao)
```

### Sobre a matriz de confusão:

Nas diagonais temos os acertos, primeiramente o acerto das transações sem fraude, depois os acertos das transações com fraudes e na outra diagonal temos os erros. Do lado direito são os erros das transações sem fraude e do lado esquerdo os erros das transações com fraude.

Basicamente, cada uma das linhas representa uma classe diferente. A primeira linha representa as transações sem fraude e a segunda linha as transações com fraude. Nós temos respectivamente quais são os acertos nas diagonais e quais que são os erros na outra diagonal.

Então a partir daqui conseguimos saber qual foi o número de acertos y_pred. Para calcularmos o número de acertos, somamos 28.419 com 36, que são a soma das transações sem fraude e com fraude. Então, com isso, chegamos no nosso resultado final, que é aquele de 0.99 que vimos antes, ou seja 99% de acerto.

Mas podemos ver que 13 das transações com fraude de 36 não representa 99%. Se formos considerar só as transações com fraude, não tivemos 99% de acerto. Se considerarmos todas as transações como a acurácia faz, tudo bem, mas eu não consigo diferenciar. Isso acontece por causa do desbalanceamento dos nossos dados.

Esse 13% e esse 36, esse pequeno erro, como estamos fazendo a função considerando o total, ele não é muito significativo, ele não tem tanto Impacto assim. Por isso é bom utilizarmos algumas outras métricas também para avaliarmos principalmente quando se trata de dados desbalanceados - o que vamos ver em seguida.

### Formulas de Acurácia, precision e recall

\begin{align}
        ACC = \frac{TP + TN}{TP + FP + TN + FN}
    \end{align}

\begin{align}
        PS = \frac{TP}{TP + FP}
    \end{align}

\begin{align}
        RC = \frac{TP}{TP + FN}
    \end{align}

Precision e recall são mais interessantes para o caso das fraudes devido ao desbalanceamento das classes positivas e negativas.

Entretanto, o recall podemos ver que não melhorou muito, na verdade até piorou um pouco. Então vamos tentar mudar mais alguns parâmetros para vermos se conseguimos melhorar esse valor de recall.

Para isso, vamos copiar de novo a nossa célula referente à execução do nosso classificador, criamos uma nova cédula. Vamos colocar mais um parâmetro, vamos digitar o min_samples_leaf=10, que vai definir qual é o número mínimo de instâncias, no nosso caso de transações, que eu tenho que ter em cada folha sempre. Vamos definir o mínimo como 10.

Ou seja, estamos forçando ele a ter pelo menos 10 instâncias nesse nó.

```python
validar_arvore(y_test, y_pred_arvore_decisao)
```

Podemos ver que o recall melhorou um pouco, mas mesmo assim, o precision piorou! Então, no caso, forçar a árvore até no mínimo 10 transações em cada nó não foi uma boa escolha para nós, então podemos ignorar esse parâmetro nesse caso específico.

Então dessas características que vimos e dessas mudanças que fizemos, o que funcionou o melhor para nós e para nossa árvore foi determinar somente a profundidade máxima igual a 10. Mas nas próximas aulas vamos ver outros algoritmos, a fim de tentarmos melhorar um pouco mais essas métricas!

## Entendendo a técnica Bagging e o Random Forest

Bagging é uma técnica de ensemble learning, que é muito popular principalmente por causa do algoritmo random forest, que vamos aprender logo mais. O bagging significa bootstrapping aggregating. 

*Bootstrapping sampling*, que é quando dividimos um grupo de instâncias em alguns grupos, algumas amostras dessas instâncias.

No nosso exemplo estamos tratando de transações, então vamos dividir esse grupo de transações em grupos menores. Esses grupos menores (amostras), têm que representar muito bem as características do nosso grupo total e também devem ser bem independentes entre si, o máximo possível.

E dizemos “o máximo possível” porque é difícil garantirmos que essa independência vai ser de 100%. Então é necessário representar 2 pontos: representar muito bem as transações do grupo maior e ser independentes entre si, o máximo possível. A ideia é que a partir de cada uma dessas amostrar nós vamos gerar um modelo, que no nosso caso é uma árvore.

Agora chegamos na parte aggregating da fórmula, que é quando vamos juntar esses modelos. Vamos agregar esses modelos em um valor único individual, que pode ser por exemplo uma soma ou uma média ou algum outro valor que represente a agregação desses modelos.

O Random Forest é um dos algoritmos mais famosos dentro da técnica bagging. Mas para entendemos como ele funciona, é importante relembrarmos como é a construção de uma árvore de decisão comum.

Para construirmos a árvore de decisão comum, fazemos isso a partir de um conjunto total, com um certo número de transações (instâncias). No caso do nosso problema, o número de instâncias exemplo é 5. Então a partir dessas 5, nós construímos uma árvore de decisão.

 Já o random forest funciona um pouco diferente: ele seleciona aleatoriamente 2 das 5 instâncias. No exemplo, selecionamos as transações 1 e 2. Ele também vai selecionar aleatoriamente alguns atributos (não todos). No nosso caso, de três atributos (pais, valor e média), ele selecionou dois: país e valor.

Então com essas duas instâncias e com esses dois atributos ele vai criar uma árvore de decisão do mesmo modo que já aprendemos. Tudo selecionando aleatoriamente.

Então, qual é o próximo passo do random forest? Ele selecionou algumas transações, selecionou alguns atributos e criou essa árvore de decisão. O próximo passo é repetir a mesma coisa: vai selecionar de novo algumas transações e alguns atributos (geralmente outras, mas ele pode repetir transações). Gerando assim mais uma arvore de decisão, e assim por diante.

Mas e agora que eu tenho duas árvores, como fazer para classificar uma nova transação que eu recebi? Eu quero classificá-la entre fraude e não-fraude. Como isso vai acontecer? Com qual das árvores eu vou me basear?

Vamos fazer a classificação para a árvore A1. Para a árvore A1 e vamos fazer da seguinte maneira: lembram que utilizamos para construir essa árvore as transações 1 e 2? Para validarmos essa árvore vamos usar só aquelas transações que NÃO usamos na construção da árvore: as transações 3, 4 e 5.

A partir dessas transações 3, 4 e 5 - que são chamadas de instâncias out of bag - vamos fazer a validação da nossa árvore e a classificação dessas transações.

Agora temos que fazer a mesma coisa para árvore A2: primeiro vamos ver quais são as instâncias out of bag delas, que são as instâncias número 1, número 2 e número 5, porque as instâncias números 3 e 4 foram aquelas que usamos na construção da árvore A2.

Depois das arvores fazerem suas respectivas classificações, fazemos uma agregação por voto. Vamos linha por linha e vendo o que tem mais votos dentro da nossa floresta aleatória.

Por exemplo: primeiro A1 não temos voto e A2 temos um voto “não”, então a maior quantidade de votos é para não.

Segunda transação: A1 não temos votos também. A2 temos o voto “não”, então um voto “não”, que é em maior quantidade (nenhum voto “sim”)- então a resposta é “não”. É a mesma coisa para transação 3 e para transação 4, as duas vão ser “não”.

 E agora podemos calcular algumas métricas para sabermos quão boa foi a classificação da nossa árvore.

Comparando com a coluna 'fraude', acertamos para transação 1 e acertamos para transação 3 e erramos para as transações 2, 4 e 5.

Então para fazermos o cálculo de uma métrica, vamos fazer 2/5 = 0.4 e agora vamos pegar o número 1 e diminuir esse 0.4. Obtemos assim o valor de 0.6, que é o erro out of bag, que é uma das métricas que podemos usar para validar quão boa é a nossa floresta aleatória.

## Entendendo o AdaBoost

Ele é como podemos ver pelo nome uma biblioteca de classificação, seu funcionamento é da seguinte maneira, ele combina diversos algoritmos de classificação fracos e transforma todos num único forte. Um algoritmo de classificação fraco é um algoritmo simples que tem um desempenho ruim, mas que funciona melhor do que ficar adivinhando respostas aleatórias.


Ele pode ser aplicado em qualquer algoritmo de classificação de dados, pois a sua técnica é baseada em muitos classificadores diferentes e não num único só, talvez ele possa te retornar um melhor resultado.


Para cada algoritmo que ele testa para classificar, com base no seu percentual o AdaBoost dá um peso para ele, por exemplo 0 se ele atingir 50% de sucesso ou se for menos o peso fica negativo, caso for maior que 50% o peso fica positivo. Algoritmos com peso negativo não são considerados na classificação pois o seus cálculos são iguais ou piores do que adivinhação para os dados testados.

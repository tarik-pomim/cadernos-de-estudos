```python
enderecos_sp = enderecos.query('sigla_uf == "SP"')
enderecos_sp
```

criando mais uma coluna rua (sempre no final a direita do DF), pegando o tipo logradouro + espaço em branco + logradouro

deixando tudo minusculo e separando com strip

```python
enderecos_sp['rua'] = enderecos_sp['tipo_logr'] + ' ' + enderecos_sp['logr_nome']
enderecos_sp['rua'] = enderecos_sp['rua'].str.lower().str.strip()
enderecos_sp.head()
```

encontraremos o problema do numero no final do endereço em dados_vendas. Extrairemos só a parte que interessa, deixando para trás numero e virgulas. Para isso foi utilizado um regex:

```python
dados_vendas['Rua'].str.extract(r'(^[\w ]+)')[:10]  # somente as 10 primeiras linhas

# ^[\w ]+ começa com tudo que for letra seguido de whitespace

# ^spam means the string must begin with spam.
# [abc] matches any character between the brackets (such as a, b, or c).
# \d, \w, and \s match a digit, word, or space character, respectively.
# The + matches one or more of the preceding group.

# fonte: automatingBoringStuffWithPython
```

Criando uma nova coluna com estes dados extraidos. Também vamos deixar tudo minusculo e tirando espaços extras com o strip:

```python
dados_vendas['apenas_rua'] = dados_vendas['Rua'].str.extract(r'(^[\w ]+)')
dados_vendas['apenas_rua'] = dados_vendas['apenas_rua'].str.lower().str.strip()
dados_vendas.head()
```

**Importante**: esta é uma das formas de fazer a vinculaçao de forma mais simples. Aqui usaremos só o nome da rua. Uma rua grande por exemplo pode ter varios ceps e cair em varias regiões censitárias. Depende de quanto detalhe vc consegue e quanto dado vc acaba perdendo por conta deste detalhamento.

Dependendo do objetivo da análise isto nao importa tanto quanto é importante ter uma base bem limpa

Tive um insight: Cruzar o dado da rua sem numero com o bairro para ter uma localização mais precisa, sem necessidade de ter o numero do imovel

Vamos fundir duas tabelas com o merge do pandas. Muitas informações serao repetidas.

```python
dados_geo = pd.merge(left=dados_vendas,  # qual base vai estar a esquerda
                     right=enderecos_sp[['rua', 'cep', 'latitude', 'longitude']],  # qual base vai estar a direita, com o fatiamento das colunas que nos interessa
                     how='left',  # quero que ela combine pela base da esquerda
                     left_on='apenas_rua',  # colunas que vao ser combinadas da esquerda
                     right_on='rua'  # colunas que vao ser combinadas da direita
                     ).drop_duplicates(subset=dados_vendas.columns).query('cep > 0')  # remove duplicados
dados_geo
```

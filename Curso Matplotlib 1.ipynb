## Linhas de restrição e anotações

```python
# adicionando linha de restrição para valores maximos e mínimos

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(df.data, df.temperatura, color='g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

eixo.axhline(max(df.temperatura), color='k', linestyle='--')  # aqui
eixo.axhline(min(df.temperatura), color='k', linestyle='--')  # aqui
```

```python
# adicionando notação

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(df.data, df.temperatura, color='g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]  # aqui
y1 = max(df.temperatura)  # aqui

eixo.annotate('Máximo', xy=(x1, y1), fontsize=20)  # aqui

eixo.axhline(max(df.temperatura), color='k', linestyle='--')
eixo.axhline(min(df.temperatura), color='k', linestyle='--')
```

```python
# adicionando a seta

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(df.data, df.temperatura, color='g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df.temperatura)  

x2 = df['data'][df['temperatura'].idxmax() - 7000]  # ajustando a palavra 'Maximo' 
y2 = max(df.temperatura) - 5  # ajustando 'maximo' mais para o centro do grafico

eixo.annotate('Máximo', xy=(x1, y1), fontsize=20, xytext=(x2, y2),
              arrowprops=dict(facecolor='k'))  # adicionando seta

eixo.axhline(max(df.temperatura), color='k', linestyle='--')
eixo.axhline(min(df.temperatura), color='k', linestyle='--')
```

```python
# adicionando a seta para minima

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(df.data, df.temperatura, color='g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df.temperatura)  
x2 = df['data'][df['temperatura'].idxmax() - 7000]  
y2 = max(df.temperatura) - 5  
eixo.annotate('Máximo', xy=(x1, y1), fontsize=20, xytext=(x2, y2),
              arrowprops=dict(facecolor='k'))  # adicionando seta

x1 = df['data'][df['temperatura'].idxmin()]  # aqui
y1 = min(df.temperatura)  
x2 = df['data'][df['temperatura'].idxmin() - 7000]  # aqui 
y2 = min(df.temperatura) + 5  
eixo.annotate('Minimo', xy=(x1, y1), fontsize=20, xytext=(x2, y2),
              arrowprops=dict(facecolor='k'))  # adicionando seta              

eixo.axhline(max(df.temperatura), color='k', linestyle='--')
eixo.axhline(min(df.temperatura), color='k', linestyle='--')
```

## Utilizando gráficos de dispersão

```python
import numpy as np
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0,0,1,1])

ax.scatter(df.comprimento_sépala, df.largura_pétala)  # aceita acento
ax.set_title('Gráfico de dispersão', fontsize=25, pad=15)
ax.set_xlabel('Comprimento da sépala', fontsize=15)
ax.set_ylabel('Largura da sépala', fontsize=15)
```

```python
# aumentando fonte dos ticks

fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0,0,1,1])

ax.scatter(df.comprimento_sépala, df.largura_pétala)
ax.set_title('Gráfico de dispersão', fontsize=25, pad=15)
ax.set_xlabel('Comprimento da sépala', fontsize=15)
ax.set_ylabel('Largura da sépala', fontsize=15)

ax.tick_params(labelsize=15)  # aqui
```

```python
# colorindo cada especie com uma cor diferente

fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g'}  # aqui

for especie in df.espécie.unique():  # este loop aqui todo
  tmp = df[df.espécie == especie]  
  ax.scatter(tmp.comprimento_sépala, tmp.largura_sépala,
             color=cores[especie])


ax.set_title('Gráfico de dispersão', fontsize=25, pad=15)
ax.set_xlabel('Comprimento da sépala', fontsize=15)
ax.set_ylabel('Largura da sépala', fontsize=15)

ax.tick_params(labelsize=15)
ax.legend(cores, fontsize=20)  # setando legenda
```

```python
# adicionando marcadores para diferenciar ainda mais os pontos

fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g'}
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor': 'o', 'Iris-virginica': 'v'}  # aqui


for especie in df.espécie.unique():
  tmp = df[df.espécie == especie]  
  ax.scatter(tmp.comprimento_sépala, tmp.largura_sépala,
             color=cores[especie],
             marker=marcadores[especie])  # aqui


ax.set_title('Gráfico de dispersão', fontsize=25, pad=15)
ax.set_xlabel('Comprimento da sépala', fontsize=15)
ax.set_ylabel('Largura da sépala', fontsize=15)

ax.tick_params(labelsize=15)
ax.legend(cores, fontsize=20)  # setando legenda
```

## Adicionando anotações e linhas horizontais

```python
# adicionando media e std como anotação

fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

mu, sigma = df.comprimento_pétala.mean(), df.comprimento_pétala.std()  # aqui

ax.hist(df.comprimento_pétala, bins=20)
ax.set_title('Histograma', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)

# aqui, os $\ faz com que apareça como fórmula pelo plt
ax.annotate(f'$\mu = {mu:.2f}$\n$\sigma={sigma:.2f}$',
            xy=(4.5, 20),
            fontsize=20)
```

```python
# adicionando uma reta para a média

fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

mu, sigma = df.comprimento_pétala.mean(), df.comprimento_pétala.std()

ax.hist(df.comprimento_pétala, bins=20)
ax.set_title('Histograma', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)

ax.annotate(f'$\mu = {mu:.2f}$\n$\sigma={sigma:.2f}$',
            xy=(4.5, 20),
            fontsize=20)

# aqui:
ax.axvline(mu, color='k', linestyle='--')
ax.annotate('Média', xy=(mu-1.3, 28), fontsize=20)  

# para os valores de xy, olha o gráfico e escolhe mais ou menos aonde quer e aí escreve na tupla
```

```python
# adicionando uma reta para a mediana

fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

mu, sigma = df.comprimento_pétala.mean(), df.comprimento_pétala.std()

ax.hist(df.comprimento_pétala, bins=20)
ax.set_title('Histograma', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)

ax.annotate(f'$\mu = {mu:.2f}$\n$\sigma={sigma:.2f}$',
            xy=(4.5, 20),
            fontsize=20)

ax.axvline(mu, color='k', linestyle='--')
ax.annotate('Média', xy=(mu-1.3, 28), fontsize=20)

# aqui
ax.axvline(df.comprimento_pétala.median(), color='g', linestyle='--')
ax.annotate('Mediana', xy=(df.comprimento_pétala.median(), 31), fontsize=20)

# para os valores de xy, olha o gráfico e escolhe mais ou menos aonde quer e aí escreve na tupla
```

```python
largura, altura = Image.open('histograma_ivc.png').size
combinada = Image.new('RGB', (3 * largura, altura))
intervalo = 0

for imagem in map(Image.open, ['histograma_iv.png', 'histograma_is.png', 'histograma_ivc.png']):
  combinada.paste(imagem, (intervalo, 0))
  intervalo += largura

combinada.save('combinada.png')
```

## Seleção das melhores variáveis com RFECV

###DecisionTreeClassifier

```python
%%time

from sklearn.feature_selection import RFECV
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

SEED = 0
np.random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y,
                                                        test_size=0.3,
                                                        random_state = SEED)

classificador_DTC = DecisionTreeClassifier(criterion='gini',
                                           max_depth=10,
                                           random_state=SEED)
classificador_DTC.fit(treino_x, treino_y)

selecionador_rfecv = RFECV(estimator=classificador_DTC,
                           cv=5,
                           step=1,
                           scoring='accuracy')
selecionador_rfecv.fit(treino_x, treino_y)

treino_rfecv = selecionador_rfecv.transform(treino_x)
teste_rfecv = selecionador_rfecv.transform(teste_x)

classificador_DTC.fit(treino_rfecv, treino_y)

y_pred = classificador_DTC.predict(teste_rfecv)
matriz_confusao = confusion_matrix(teste_y, y_pred)

plt.figure(figsize = (10, 8))
sns.set(font_scale = 2)

ax = sns.heatmap(matriz_confusao, annot=True, fmt='d')
ax.set(xlabel = 'Predição', ylabel = 'Real')

print(f'Acurácia: {classificador_DTC.score(teste_rfecv, teste_y) * 100 :.2f}%')
```

Mas como interpretá-la? O eixo Y está representando nossos valores reais, e o eixo X representa os valores de predição. Temos 0 quando é churn-não e 1 quando o é churn-sim. A soma dos dois quadrados superiores nos trará o total de churns-não, e a dos dois quadrados inferiores o total de churn-sim.

De 1522 churn-não, nosso modelo acertou 1214, classificando o restante (302) como churn-sim. Já nos casos de churn-sim, nosso modelo acertou 1333, classificando 243 como churn-nao. Acuracia de 82.21%

### Random Forest Classifier

```python
%%time

from sklearn.feature_selection import RFECV
from sklearn.model_selection import train_test_split
from sklearn.tree import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

SEED = 0
np.random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y,
                                                        test_size=0.3,
                                                        random_state = SEED)

classificador_RFC = RandomForestClassifier(n_estimators=100,
                                           random_state=SEED)
classificador_RFC.fit(treino_x, treino_y)

selecionador_rfecv = RFECV(estimator=classificador_RFC,
                           cv=5,
                           step=1,
                           scoring='accuracy')
selecionador_rfecv.fit(treino_x, treino_y)

treino_rfecv = selecionador_rfecv.transform(treino_x)
teste_rfecv = selecionador_rfecv.transform(teste_x)

classificador_RFC.fit(treino_rfecv, treino_y)

y_pred = classificador_RFC.predict(teste_rfecv)
matriz_confusao = confusion_matrix(teste_y, y_pred)

plt.figure(figsize = (10, 8))
sns.set(font_scale = 2)

ax = sns.heatmap(matriz_confusao, annot=True, fmt='d')
ax.set(xlabel = 'Predição', ylabel = 'Real')

print(f'Acurácia: {classificador_RFC.score(teste_rfecv, teste_y) * 100 :.2f}%')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

def classificar_KNN_TTS(x):
  SEED = 0
  np.random.seed(SEED)

  treino_x, teste_x, treino_y, teste_y = train_test_split(x, y,
                                                          test_size=0.3,
                                                          random_state = SEED)
  
  classificador_KNN = KNeighborsClassifier(metric='euclidean')
  classificador_KNN.fit(treino_x, treino_y)

  predito_knn = classificador_KNN.predict(teste_x)

  print(f'Acurácia KNN/TTS: {classificador_KNN.score(teste_x, teste_y) * 100 :.2f}%')
  
```

### Bernoulli Naive Bayes com Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
import numpy as np

def classificar_BNB_TTS(x):
  SEED = 0
  np.random.seed(SEED)

  mediana = np.median(x)

  treino_x, teste_x, treino_y, teste_y = train_test_split(x, y,
                                                          test_size=0.3,
                                                          random_state = SEED)
  
  classificador_BNB = BernoulliNB(binarize=mediana)
  classificador_BNB.fit(treino_x, treino_y)

  print(f'Acurácia BNB/TTS: {classificador_BNB.score(teste_x, teste_y) * 100 :.2f}%')
  
```

```python
classificar_BNB_TTS(x_normalizado)
```

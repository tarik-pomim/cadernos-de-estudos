## <font color='red'>Problema</font>

Você é um pesquisador que estuda o mercado de trabalho brasileiro e resolve estudar as diferenças salariais dos trabalhadores dos estados do Rio de Janeiro e de São Paulo. Durante sua pesquisa você verifica que, aparentemente, os rendimentos dos trabalhadores no estado do Rio de Janeiro são mais baixos que os rendimentos dos trabalhadores no estado de São Paulo. Para confirmar esta conclusão realize um teste de hipótese de comparação de médias em cima de duas amostras de trabalhadores dos dois estados. Siga as seguintes etapas:

- Selecione duas amostras de **500 trabalhadores** para cada um dos dois estados (variável UF) para realizar o teste. Utilize o **parâmetro random_state=101**.
- Considere o **nível de significância de 5%**.
- Teste a hipótese de que a **renda média dos trabalhadores do Rio de Janeiro é menor que a renda média dos trabalhadores de São Paulo**.

---

### Seleção das amostras

```python
renda_trabalhadores_RJ = dados.query('UF == 33 and Renda > 0').sample(n=500, random_state=101).Renda
renda_trabalhadores_RJ.head()
```

```python
renda_trabalhadores_SP = dados.query('UF == 35 and Renda > 0').sample(n=500, random_state=101).Renda
renda_trabalhadores_SP.head()
```

### Dados do problema
Obter média e desvio-padrão para as duas amostras

```python
media_amostra_RJ = renda_trabalhadores_RJ.mean().round(2)
media_amostra_RJ
```

```python
std_amostra_RJ = renda_trabalhadores_RJ.std().round(2)
std_amostra_RJ
```

```python
media_amostra_SP = renda_trabalhadores_SP.mean().round(2)
media_amostra_SP
```

```python
std_amostra_SP = renda_trabalhadores_SP.std().round(2)
std_amostra_SP
```

```python
significancia = 0.05
confianca = 1 - significancia
n_RJ = 500
n_SP = 500
D_0 = 0
```

### **Passo 1** - formulação das hipóteses $H_0$ e $H_1$

#### <font color='red'>Lembre-se, a hipótese nula sempre contém a alegação de igualdade</font>

Teste a hipótese de que a renda média dos trabalhadores do Rio de Janeiro é menor que a renda média dos trabalhadores de São Paulo.

H0: u1 >= u2

H1: u1 < u2


H0: u1 - u2 >= 0

H1: u1 - u2 > 0


---

### **Passo 2** - escolha da distribuição amostral adequada

### O tamanho da amostra é maior que 30?
#### Resp.: Sim

### Podemos afirmar que a população se distribui aproximadamente como uma normal?
#### Resp.: Não

### O desvio padrão populacional é conhecido?
#### Resp.: Não

---

### **Passo 3** - fixação da significância do teste ($\alpha$)

```python
from scipy.stats import norm
```

```python
probabilidade = significancia
probabilidade
```

```python
z_alpha = norm.ppf(probabilidade)
z_alpha.round(2)
```

---

### **Passo 4** - cálculo da estatística-teste e verificação desse valor com as áreas de aceitação e rejeição do teste

```python
numerador = (media_amostra_RJ - media_amostra_SP) - D_0

denominador = np.sqrt((std_amostra_RJ ** 2 / n_RJ) + (std_amostra_SP ** 2 / n_SP))

z = numerador / denominador
z
```

---

### **Passo 5** - Aceitação ou rejeição da hipótese nula

### <font color='red'>Critério do valor crítico</font>

```python
z <= z_alpha  # rejeitamos H0 if true
```

### <font color='red'>Critério do valor $p$</font>

### Utilize DescrStatsW

```python
from statsmodels.stats.weightstats import DescrStatsW, CompareMeans
```

```python
test_RJ = DescrStatsW(renda_trabalhadores_RJ)
```

```python
test_SP = DescrStatsW(renda_trabalhadores_SP)
```

```python
test = test_RJ.get_compare(test_SP)
```

```python
z, p_valor = test.ztest_ind(alternative='smaller', value=0)
```

```python
print(f'Z = {z.round(2)}')
print(f'P_valor = {p_valor.round(2)}')
```

```python
p_valor <= significancia  # rejeitamos h0 if true
```

### Utilize CompareMeans

```python
test = CompareMeans(test_RJ, test_SP)
test = test_RJ.get_compare(test_SP)
z, p_valor = test.ztest_ind(alternative='smaller', value=0)
p_valor.round(2)
```

```python
p_valor <= significancia  # rejeita H0 if true
```

### <font color='green'>Conclusão: Com um nível de confiança de 95% rejeitamos H0, isto é, concluímos que a média das rendas dos moradores do RJ é menor que a média das rendas dos moradores de SP (considerando rendas de R$0 na conta).</font>

Mann-Whitney: Utilizado para verificar se duas amostras independentes foram selecionadas a partir de populações que têm a mesma média;

Wilcoxon: Empregado quando se deseja comparar duas amostras relacionadas, amostras emparelhadas. Pode ser aplicado quando se deseja testar a diferença de duas condições, isto é, quando um mesmo elemento é submetido a duas medidas;

Qui-quadrado: Também conhecido como teste de adequação ao ajustamento. Testa a hipótese nula de não haver diferença entre as frequências observadas de um determinado evento e as frequências que são realmente esperadas para este evento
